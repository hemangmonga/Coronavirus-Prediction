```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
df = pd.read_csv('IndividualDetails.csv')
df = df.drop(['id', 'government_id', 'detected_city', 'nationality', 'notes', 'status_change_date', 'age', 'gender'], axis=1)
df['diagnosed_date'] = pd.to_datetime(df['diagnosed_date'], dayfirst = True)
df
```

```python
States = list()
for i in df['detected_state'].unique():
    States.append(i)
States.remove('Mizoram')
States.remove('Arunachal Pradesh')
# States.remove('Dadra and Nagar Haveli')
States_df = States.copy()
df = df.set_index('detected_state')   
df
```

```python
district_list = list()
for i in States:
    district_list.append(i)
    temp = df.loc[i]
    district_list.append(temp['detected_district'].value_counts())
district_list[21]
```

```python
#To Create data for plotting Map
#To get data for different states, Change no. in District_list['here']
# No. can only be changed in odd nos. example 1,3,5,etc
# We you change in even no, it will give errors.
# 1 for Kerala, 3 for Delhi.
Rajasthan = pd.DataFrame(district_list[7]).reset_index().rename(columns = {'index':'DIST_NAME', 'detected_district':'Count'})#.drop(21)
Rajasthan.to_csv('Rajasthan.csv')
```
